```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
df = pd.read_csv('./movies.csv')
```

```python
df.isnull().sum()
```

```python
df['genres'].astype(str)
```

```python
df['genres'] = df['genres'].str.split('|')

df
```

```python
mlb = MultiLabelBinarizer()
generos_binarios = mlb.fit_transform(df['genres'])
df_generos = pd.DataFrame(generos_binarios, columns=mlb.classes_)

df_generos
```

```python
kmeans = KMeans(n_clusters=10, random_state=0)
clusters = kmeans.fit_predict(df_generos)

df['Cluster'] = clusters
```

```python
filme_escolhido = 'Father of the Bride Part II (1995)'
cluster_filme = df.loc[df['title'] == filme_escolhido, 'Cluster'].values[0]
recomendacoes = df[df['Cluster'] == cluster_filme].loc[df['title'] != filme_escolhido]

recomendacoes[['title', 'genres']]
```
