```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
datas = {
    'Idade': [37, 46, 62, 50, 61, 50, 54, 27, 31, 56],
    'Casado': ['SIM', 'NÃO', 'SIM', 'SIM', 'SIM', 'SIM', 'NÃO', 'NÃO', 'NÃO', 'SIM'],
    'Filhos': [2, 0, 1, 2, 3, 2, 0, 0, 0, 3],
    'Financiamento': ['NÃO', 'SIM', 'NÃO', 'SIM', 'SIM', 'NÃO', 'SIM', 'NÃO', 'NÃO', 'SIM'],
    'Rendimentos': [17729.80, 41016.00, 26909.20, 22522.80, 57880.70, 16497.30, 38446.60, 15538.00, 12640.30, 41034.00],
    'Sexo': ['M', 'F', 'M', 'M', 'F', 'M', 'M', 'M', 'M', 'F']
}

df =pd.DataFrame(datas)
```

```python
df
```

```python
for index, row in df.iterrows():
    if row['Casado'] == 'SIM':
        df.loc[index, 'Casado'] = 1
    else: 
        df.loc[index, 'Casado'] = 1

    if row['Financiamento'] == 'SIM':
        df.loc[index, 'Financiamento'] = 1
    else: 
        df.loc[index, 'Financiamento'] = 1

    if row['Sexo'] == 'M':
        df.loc[index, 'Sexo'] = 1
    else: 
        df.loc[index, 'Sexo'] = 1
```

```python
df
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['Rendimentos']])
transform = scaler.transform(df[['Rendimentos']])

df['Rendimentos'] = transform
```

```python
df
```

```python
datas2 = {
    'valores': [3,9,5,11,5,7]
}
df2 = pd.DataFrame(datas2)

scaler.fit(df2[['valores']]) 
transform3 = scaler.transform(df2[['valores']])

df2['Re-escala']=transform3

scaler2 = StandardScaler()
scaler2.fit(df2[['valores']]) 
transform4 = scaler2.transform(df2[['valores']])

df2['Padronizacao']=transform4
```
