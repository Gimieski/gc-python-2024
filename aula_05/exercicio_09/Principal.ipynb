```python
import pandas as pd

filmes = pd.read_json('filmes_preferidas.json')
animes = pd.read_json('animes_preferidas.json')
series = pd.read_json('series_preferidas.json')

novos_filmes = pd.DataFrame({
    'filmes': ['Matrix', 'O Senhor dos Anéis', 'Harry Potter', 'Batman: O Cavaleiro das Trevas', 'Avatar'],
    'ano': [1999, 2001, 2002, 2008, 2009],
    'nota': [9, 10, 6, 7, 4]
})

novos_animes = pd.DataFrame({
    'animes': ['Attack on Titan', 'Death Note', 'One Piece', 'Naruto', 'Dragon Ball Z'],
    'ano': [2013, 2006, 1999, 2002, 1989],
    'nota': [10, 7, 4, 8, 10]
})

novos_series = pd.DataFrame({
    'series': ['Breaking Bad', 'Game of Thrones', 'Sherlock', 'Friends', 'The Office'],
    'ano': [2008, 2011, 2010, 1994, 2005],
    'nota': [10, 5, 8, 6, 10]
})

# adicionei mais linhas
animes = pd.concat([animes, novos_animes], ignore_index=True)
filmes = pd.concat([filmes, novos_filmes], ignore_index=True)
series = pd.concat([series, novos_series], ignore_index=True)
```

```python
# maior avaliados
animes_maior_avaliados= animes.loc[animes['nota'] >= 9]
filmes_maior_avaliados=filmes.loc[filmes['nota'] >= 9]
series_maior_avaliados=series.loc[series['nota'] >= 9]
```

```python
listagem_concatenada=pd.concat([animes, filmes, series], axis=1)

# 3 primeiras colunas
listagem_concatenada.iloc[:, :3]

# 10 primeiras linhas
listagem_concatenada.head(10)
```

```python
# valores maiores que meida
animes_maior_media = animes.loc[animes['nota'] > animes['nota'].mean()]
filmes_maior_media = filmes.loc[filmes['nota'] > filmes['nota'].mean()]
series_maior_media = series.loc[series['nota'] > series['nota'].mean()]
```
